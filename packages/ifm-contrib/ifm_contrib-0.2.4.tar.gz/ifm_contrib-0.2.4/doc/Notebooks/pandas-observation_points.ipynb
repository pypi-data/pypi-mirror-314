# doc.c.obs.gdf - Observation Points to GeoDataFrames

The `doc.c.obs.gdf` object provides functions to obtain the location of observation points as a GeoDataframe (2D and 3D Layered only).

GeoDataFrames are objects similar to shapefiles. They can be processed using GeoPANDAS, a Python library with usefull features for 2D Geometry operations including export to shp-files and basic GIS operations. See the project website [geopandas.org](https://www.geopandas.org) for further information.

*contact: [Alex Renz](https://github.com/red5alex), are@dhigroup.com* 

```python
import sys
sys.path.append("C:/Program Files/DHI/2019/FEFLOW 7.2/python")
import ifm_contrib as ifm 
ifm.forceLicense("Viewer")
doc = ifm.loadDocument("./../../unittests/models/example_2D.dac")
```

`doc.c.obs.gdf.obspoints()` returns a geodataframe with all data stored in the FEFLOW model.

```python
doc.c.obs.gdf.obspoints()
```

Plotting is most usefull when combined with other GeoPandas objects of the model. **Note** that the `doc.c.plot` module provides easier accessible functions plotting (and not processing) is the single puporse.

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.axis('equal')
ax1 = plt.gca()  # get current axis
doc.c.mesh.gdf.elements().plot(ax=ax1)
doc.c.obs.gdf.obspoints().plot(ax=ax1, color="red")
# add labels
for label, row in doc.c.obs.gdf.obspoints().iterrows():
    plt.annotate(s=label, 
                 xy=(row['x'], row['y']),
                 horizontalalignment='center')
```

Note: `doc.c.plot.obs_markers` and `doc.c.plot.obs_labels` are a more efficient way of plotting observation points on a map.
